```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
data=pd.read_csv(r"C:\Users\Ishan\Downloads\IPL-First-Innings-Score-Prediction-Deployment-master\IPL-First-Innings-Score-Prediction-Deployment-master\ipl.csv")
data
```

```python
data.drop(["mid","venue","batsman","bowler","striker","non-striker"],axis=1,inplace=True)
data
```

```python
data["bat_team"].unique()
```

```python
teams=['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians', 'Kings XI Punjab',
       'Royal Challengers Bangalore', 'Delhi Daredevils','Sunrisers Hyderabad']
```

```python
data=data[(data["bat_team"].isin(teams)) & (data["bowl_team"].isin(teams))]
data
```

```python
data=data[data["overs"]>=5.0]
```

```python
data["date"]=pd.to_datetime(data["date"],format="%Y-%m-%d")
data.dtypes
```

```python
data=pd.get_dummies(data=data,columns=["bat_team","bowl_team"])
data
```

```python
data.dtypes
```

```python
data.columns
```

```python
data=data[['date','bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils',
       'bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils',
       'bowl_team_Kings XI Punjab', 'bowl_team_Kolkata Knight Riders',
       'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
       'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad','runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5',
       'total']]
```

```python
data
```

```python
X_train =data.drop(labels='total', axis=1)[data['date'].dt.year <= 2016]
X_test = data.drop(labels='total', axis=1)[data['date'].dt.year >= 2017]

y_train = data[data['date'].dt.year <= 2016]['total'].values
y_test = data[data['date'].dt.year >= 2017]['total'].values
```

```python
X_train
```

```python
X_train.drop(labels='date', axis=True, inplace=True)
X_test.drop(labels='date', axis=True, inplace=True)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={"alpha":np.arange(0.01,1,0.1)}
```

```python
r_grid=GridSearchCV(ridge,param_grid=param_grid,scoring="neg_mean_squared_error",cv=5)
r_grid.fit(X_train,y_train)
```

```python
r_grid.best_score_
```

```python
import xgboost
xgb=xgboost.XGBRegressor()
```

```python

xgb.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
y_pred=xgb.predict(X_test)
```

```python
mean_absolute_error(y_pred,y_test)
```
